### useReducer
> 是 React 内置 Hook，用来管理复杂状态，是 useState 的「升级版」。

* useState：简单状态（单个值、少量数据）
* useReducer：多字段、关联状态、复杂修改逻辑、表单、购物车、计数器多操作
* 一句话：状态逻辑多、改法多，就用 useReducer
2. 核心公式
jsx
const [state, dispatch] = useReducer(reducer, 初始值)
三部分：
state：当前所有状态（只读）
dispatch：发「动作指令」给 reducer
reducer：纯函数，接收 (当前state, action) → 返回新 state
- 不能返回原变量我的profile， 系统新旧值比较没变化，不会更新。
```
 const [我的profile, 干活] = useReducer((我的profile, action) => {
    switch (action.type) {
      case "age":

         return { ...我的profile,age: 我的profile.age +1};
      case "name":
        
         return { ...我的profile,name: 我的profile.name + "!" };

      default:
        return 我的profile;
    }
  }, { name: "Jack", age: 18 });
```
