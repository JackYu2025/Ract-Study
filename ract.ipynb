- StrictMode 严格模式

> StrictMode 是 React 自带的开发环境检查工具，只在开发模式生效，生产环境会自动忽略，不会影响运行速度。它的作用就是：提前帮你找出代码里的潜在 Bug、过时 API、不安全写法

> 要不要删掉 StrictMode？强烈建议保留 只在开发环境生效 能提前发现大量 Bug StrictMode 是「运行时检测」，不是编辑器静态检查
