
## ***useReducer***
> 用来管理复杂状态，是 useState 的「升级版」。

* useState：简单状态（单个值、少量数据）
* useReducer：多字段、关联状态、复杂修改逻辑、表单、购物车、计数器多操作
* 一句话：状态逻辑多、改法多，就用 useReducer

const [state, dispatch] = useReducer(reducer, 初始值)

* 三部分：
> state：当前所有状态（只读）
> dispatch：发「动作指令」给 reducer
> reducer：纯函数，接收 (当前state, action) → 返回新 state, reducer()的***state*** 没有必要和state同名

- 不能返回原变量我的profile， 系统新旧值比较没变化，不会更新。

* 定义
```
 const [我的profile, 干活] = useReducer((我的profile, action) => {
    switch (action.type) {
      case "age":
         return { ...我的profile,age: 我的profile.age +1};
      case "name":       
         return { ...我的profile,name: 我的profile.name + "!" };
      default:
        return 我的profile;
    }
  }, { name: "Jack", age: 18 });
```


## ***useRef***
对比 useRef vs useState
* useState：变了 → 重新渲染   
* useRef：变了 → 不渲染，只存

useRef = 组件内的 “全局变量”+ DOM 抓手，改 current 不刷新页面。

***核心作用***   
保存 DOM 元素 / 组件实例（直接操作 DOM）
保存变量， 渲染不变、修改不触发重渲染

* 定义
```
  const sound =useRef();
  const color =useRef();
```
* 使用
```
  const submit = (e0bj) => {
    e0bj.preventDefault();
    const soundValue = sound.current.value;
    const colorValue = color.current.value;
    console.log(`This is text: ${soundValue}`);
    console.log(`This is color: ${colorValue}`);
  };
  return (
    <form onSubmit={submit}>
      <input type="text" placeholder="Sound..." ref={sound} />
      <input type="color" placeholder="Enter your favorite color" ref={color} />
      <button type="submit">Add</button>
    </form>

  );
```

* {...title} —— 对象展开运算符
>  会把这个对象里所有键值对，自动拆分成 HTML 属性 赋给网页element 像 <input>，相当于<input ...> 里的属性对。

* 定义 文件useInput.js
  ```
  import { useState } from "react";
  export function useInput(initialValue) {
    const [value, setValue] = useState(initialValue);
    return [
      { value, onChange: (e) => setValue(e.target.value) },
      () => setValue(initialValue),
    ];
  }
  ```

# Fetch
```
export function useFetch(uri) {
    const [data, setData] = useState();
    const [loading, setLoading] = useState(true);
    const [error, setError] = useState();

    useEffect(() => {
        if (!uri) return;
        fetch(uri)
            .then((data) => data.json())
            .then(setData)
            .then(() => setLoading(false))
            .catch(setError);
    }, [uri]);

    return { loading, data, error };
}
```
